```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("results.csv")
df
```

```python
df["comprehension"] = df[["q1", "q2", "q3"]].mean(axis=1)
df
```

```python
summary = df.groupby("condition")["comprehension"].mean()
summary
```

```python
summary.plot(kind="bar")
plt.ylabel("Mean comprehension score")
plt.title("Baseline vs X-SHIELD comprehension")
plt.tight_layout()
plt.savefig("../paper/figures/comprehension_baseline_vs_xshield.png")
plt.show()
```
